```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_std=sc.fit_transform(x)
x_std=pd.DataFrame(x_std)
```

```python
from sklearn.cluster import AgglomerativeClustering

inertia = []
for aff in ['euclidean','manhattan','cosine']:
    for lk in ['single','complete','average']:
        model1 = AgglomerativeClustering(n_clusters=3,affinity=aff,linkage=lk)
        model1.fit(x_std)
        L=pd.DataFrame(model1.labels_)
        df_labels = x_std.copy(deep=True) 
        df_labels['labels']=model1.labels_
        df0=df_labels[df_labels['labels'] == 0]
        df1=df_labels[df_labels['labels'] == 1]
        df2=df_labels[df_labels['labels'] == 2]
        c0=np.array(np.mean(df0.iloc[:,:-1]))
        c1=np.array(np.mean(df1.iloc[:,:-1]))
        c2=np.array(np.mean(df2.iloc[:,:-1]))
        i1=0
        i2=0
        i3=0
        for i in np.arange(df0.shape[0]):
            i1 = i1 + np.sum((df0.iloc[i,:-1]-c0)**2)
        for i in np.arange(df1.shape[0]):
            i2 = i2 + np.sum((df1.iloc[i,:-1]-c1)**2)
        for i in np.arange(df2.shape[0]):
            i3 = i3 + np.sum((df2.iloc[i,:-1]-c2)**2)
        inertia.append(i1+i2+i3)
inertia
```

# APPLYING PCA

```python
#attach the cluster or groups created to the original scaled data frame
df_with_label = x_std.copy(deep = True)
df_with_label['labels'] = model.labels_

df_with_label.head()
```

```python
x_std=df_with_label.drop('labels',axis=1)
y=df_with_label['labels']
```

```python
cov_matrix=np.cov(x_std.T)           
cov_matrix_df=pd.DataFrame(cov_matrix)
cov_matrix_df
```

```python
eig_vals,eig_vect=np.linalg.eig(cov_matrix)
print('eigen vectors \n%s',eig_vect)
print('eigen values \n%s',eig_vals)
```

```python
tot=sum(eig_vals)
var_exp=[(i/tot)*100 for i in sorted(eig_vals,reverse=True)]
cum_var_exp=np.cumsum(var_exp)
print('cummulative varience explained',cum_var_exp)
print('individua varience explained  ',var_exp)
```

```python
#graphically checking the cummilative and indiavidual varience
plt.figure(figsize = (6,4))
plt.bar(range(x_std.shape[1]), var_exp, alpha = 0.5, align='center', label='Individual explained variance')
plt.step(range(x_std.shape[1]), cum_var_exp, where = 'mid', label='Cumulative explained Variance')
plt.ylabel('Explained Variance Ratio')
plt.ylabel('Principal Components')
plt.legend(loc= 'best')
plt.tight_layout()
plt.show()
```

```python
 eig_vals,eig_vect=np.linalg.eig(cov_matrix)
print('eigen vectors \n%s',eig_vect)
print('eigen values \n%s',eig_vals)
```

```python
eig_pairs=[(eig_vals[index],eig_vect[:,index])for index in range(len(eig_vals))]
```

```python
eig_pairs.sort()
eig_pairs.reverse()
```

```python
#extracting the descending ordered eigen values and eigen vectors seperately
eig_vals_sort=[eig_pairs[index][0] for index in range (len(eig_vals)) ]
eig_vect_sort=[eig_pairs[index][1] for index in range (len(eig_vals)) ]
```

```python
print(eig_vals_sort)
```

```python
#converting  eig_vect_sort from list to numpy array column wise to perform dot product with scaled data
eig_vect_sorted=np.array(eig_vect_sort).T
eig_vect_sorted
```

```python
eig_vect_sorted.shape
```

```python
pc_reduce=eig_vect_sorted[:,:3]
x_pca=np.dot(x_std,pc_reduce)
```

```python
x_pca.shape
```

```python
from sklearn import model_selection
from sklearn.model_selection import KFold, cross_val_score
import matplotlib.pyplot as plt
results = []
names = []
models=[]
models.append(('Knn_base',knn_base_model))
# models.append(('knn_bagges_model',knn_bag_model))

for name, model in models:
    kfold = model_selection.KFold(shuffle = True, n_splits = 3, random_state = 0)
    cv_results = model_selection.cross_val_score(model,x_pca,y,cv = kfold, scoring='f1_weighted')
    results.append(cv_results)
    names.append(name)
    print('NAME: {} ----- ACCURACY: {} ----- VAR: {} '.format(name,np.mean(cv_results),np.var(cv_results,ddof=1)))
    
    #boxplot algo comparison
fig = plt.figure()
fig.suptitle('Algo Comparison')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```
